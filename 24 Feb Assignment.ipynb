#Q10 ANSWER

```python
dt=pd.date_range(start="2023-06-01",end="2023-06-30")
```

```python
sales=list(range(10,100,3))
```

```python
d={
    "date":dt,
    "sales":sales
}
```

```python
df=pd.DataFrame(d)
```

```python
df
```

```python
def movingaveragecolumn(df):
    df["MovingAverage"]=df.sales.rolling(window=7).mean()
    return df
```

#Q11 ANSWER

```python
def function1(x):
    x["weekday"]=x["Date"].dt.weekday
    def weekdayfunction(x):
        if x==0:
            return "Monday"
        elif x==1:
            return "Tuesday"
        elif x==2:
            return "Wednesday"
        elif x==3:
            return "Thursday"
        elif x==4:
            return "Friday"
        elif x==5:
            return "Saturday"
        else:
            return "Sunday"
    x["weekday"]=x.weekday.apply(weekdayfunction)
    return x
```

```python
df=pd.DataFrame()
```

```python
df["Date"]=pd.date_range(start="2023-01-01",end="2023-01-31")
```

```python
df
```

```python
function1(df)
```
